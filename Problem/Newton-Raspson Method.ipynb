# Giải hệ phương trình phi tuyến tính 

## 1. Đặt vấn đề

Tìm nghiệm là phần quan trọng trong lĩnh vực đại số.

Phương trình tuyến tính (Linear Equation) là dạng thường gặp, dưới đây là một số ví dụ phương trình tuyến tính và hệ phương trình tuyến tính.



*   $2x + 1 = -2$

*   $\begin{cases} 3x + 5y =0  \\ -6x + y = -2 \end{cases}$

*   $\begin{cases} 3x + 5y + z = -1 \\ 7x – 2y + 4z = 0 \\ -6x + 3y + 2z  = 1\end{cases}$




Các hệ phương trình tuyến tính này có thể giải bằng phương pháp [Gauss-Jordan](https://www.mathsisfun.com/algebra/matrix-inverse-row-operations-gauss-jordan.html) hoặc bằng phương pháp [Crammer](https://math.libretexts.org/Bookshelves/Precalculus/Book%3A_Precalculus_(OpenStax)/09%3A_Systems_of_Equations_and_Inequalities/9.09%3A_Solving_Systems_with_Cramer's_Rule).

- Thuật toán Gauss-Jordan

        1.   Từ hệ phương trình tuyến tính đã cho chuyển về ma trận mở rộng.

        2.   Dùng các phép biến đổi sơ cấp trên ma trận để đưa ma trận về dạng ma trận bậc thang.
        
        3.   Biện luận nghiệm của hệ đã cho dựa trên ma trận bật thang thu được.


- Thuật toán Crammer






Dưới đây là một hệ phương trình 3 ẩn $x, y, z$ được giải bằng phương pháp Gauss-Jordan.

$$\begin{cases} 3x + 5y + z = -1 \\ 7x – 2y + 4z = 0 \\ -6x + 3y + 2z  = 1\end{cases}$$

1. Chuyển hệ phương trình tuyến tính về ma trận mở rộng
$$ \~A = 
\left(
\begin{array}{ccc|c}
3  & 5 & 1 & -1\\
7 & 2 & 4 & 0 \\
-6 & 3 & 2 & 1 \\
\end{array}
\right)
$$

2. Thực hiện các phép biến đổi sơ cấp trên ma trận để đưa ma trận mở rộng về dạng bậc thang.

$$ \~A = 
\left(
\begin{array}{ccc|c}
3  & 5 & 1 & -1\\
7 & 2 & 4 & 0 \\
-6 & 3 & 2 & 1 \\
\end{array}
\right)
$$

$$ \rightarrow \left(
\begin{array}{ccc|c}
3  & 5 & 1 & -1\\
0 & \frac{-29}{3} & \frac{5}{3} & \frac{7}{3} \\
0 & 13 & 4 & -1 \\
\end{array}
\right) $$

$$ \rightarrow \left(
\begin{array}{ccc|c}
3  & 5 & 1 & -1\\
0 & \frac{-29}{3} & \frac{5}{3} & \frac{7}{3} \\
0 & 0 & \frac{181}{29} & \frac{62}{29} \\
\end{array}
\right) $$

3. Biện luận giải hệ phương trình dựa trên ma trận bậc thang

   Hạng của ma trận A bằng với hạng của ma trận mở rộng $\~A$, $r(A) = r(\~A)$ nên hệ phương trình tuyến tính đã cho có nghiệm duy nhất.

$$\begin{cases} 
    x = (-1 - z - 5y) \div 3 = \frac{-26}{181} \\ 
    y = (\frac{7}{3} - \frac{5}{3}z) \div \frac{-29}{3} = \frac{-33}{181} \\ 
    z = \frac{62}{29} \div \frac{181}{29} = \frac{62}{181}
\end{cases}$$

Tuy nhiên có nhiều bài toán trong thực tế thường tồn tại dưới dạng tập hợp các phương trình phi tuyến, ví dụ như một hệ phương trình phi tuyến dưới đây.




$$\begin{cases} 
   3x_1 - cos(x_2 x_3) - \frac{1}{2} = 0 \\
   x_1^2  - 81 (x_2 + 0.1)^2 + sin(x_3) + 1.06 = 0 \\
   e^{x_1x_2} + 20x_3 + \frac{10 \pi - 3 }{3} = 0
\end{cases}$$

Đương nhiên không thể dùng các phương pháp Gauss-Jordan hay Crammer để giải. Ở đây ta có thể dùng phương pháp Newton-Raspson để tìm ra nghiệm xấp xỉ của hệ phương trình phi tuyến trên một cách dễ dàng và hoàn toàn có thể đảm bảo việc nghiệm tìm được có gần với nghiệm thực sự nhất.

## 2. Phương pháp Newton-Raspson

### Xác định Input và Output

- Input: 
    + n phương trình (equatiosn) và các ẩn (unkowns/variables)
    + $x = (x_1, x_2, ..., x_n)^t$ là vector nghiệm khởi tạo.
    + Epxilon: sai số cho phép
    + N: số lượng vòng lặp tối đa.

- Output:
    + Nghiệm xấp xỉ $x = (x_1, x_2, ..., x_n)^t$

### Mô phỏng thuật toán dưới dạng mã giả

1. Set k = 1
2. While ( $k \le N$) do

    2.1 Calculate the $F(x)$ and $J(x)$.

    2.2 Solve the $n \times n$ linear system J(x)y = -F(x)
    
    2.3 Set x = x + y
    
    2.4 if $||y|| < EPXILON$ then 
                return x // Successfull
    2.5 k = k + 1

3. Return x 

    // Unsuccessfull because Maximum number of iterators exceeded


Phân tích
- Thuật toán Newton-Raspson nhận đầu vào là một hệ phương trình phi tuyến/tuyến tính, giá trị nghiệm khởi tạo, sai số cho phép, số lượng vòng lặp tối đa.

- Thuật toán sẽ dừng trong hai trường hợp:
    + Trường hợp tìm được vector nghiệm thỏa mãn sai số cho phép (bước 2.4)
    + Trường hợp đã lặp hết N lần nhưng không thể tìm được nghiệm thỏa mãn sai số cho phép.

### Coding

```
# Khai báo các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
```

```
# Định nghĩa hệ phương trình
def F(x: np.ndarray):
    f1 = 3*x[0] - np.math.cos(x[1]*x[2]) - 1/2
    f2 = x[0]**2 -81*(x[1]+0.1)**2 + np.math.sin(x[2]) + 1.06
    f3 = np.e**(-x[0]*x[1]) + 20*x[2] + (10*np.pi-3)/3
    return np.array([f1, f2, f3])
```

```
# Định nghĩa đạo hàm của hệ phương trình: ma trận Jacobian
def J(x: np.ndarray):
    J1 = np.array([3, x[2]*np.sin(x[2]*x[1]), x[1]*np.sin(x[1]*x[2])])
    J2 = np.array([2*x[0], -162*(x[1]+0.1), np.cos(x[2])])
    J3 = np.array([-x[1]*np.e**(-x[0]), -x[0]*np.e**(-x[1]), 20])
    return np.array([J1, J2, J3])
```

```
# Định nghĩa thuật toán Newton-Raspson cho bài toán
def solve(x: np.ndarray, epxilon: float, N: int):
    k = 1
    while k < N:
        Fx = F(x=x)  
        Jx = J(x=x)
        y = np.linalg.solve(Jx, Fx)
        x = x - y
        if np.linalg.norm(y, ord=np.inf) <= epxilon:
            return x, k
        k = k + 1

    return x, k-1
```

```
# Giải hệ phương trình bằng việc gọi hàm
x = np.array([0.1, 0.1, -0.1])   # Khởi tạo giá vector nghiệm
epxilon = float(input("Giá trị sai số cho phép epxilon ? "))
N = int(input("Nhập số lần lặp giới hạn cho phép N ? "))

x, n = solve(x=x, epxilon=epxilon, N=N)

if n > N:
    print("Không tìm được nghiệm đảm bảo sai số cho trước trong {0} vòng lặp".format(N))
else:
    np.set_printoptions(precision=6)
    print("Nghiệm tìm được trong {0} vòng lặp và có giá trị là {1}".format(n, x))
```
